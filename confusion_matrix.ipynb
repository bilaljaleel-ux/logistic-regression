```python
#This notebook explains the concept of a confusion matrix and demonstrates
#how it is used to evaluate classification models using a loan default example.
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"D:\Hands on Stat\optiva (1).xlsx")
df.head()
```

```python
#-Actual values represent the true loan default status.
#- Predicted values represent the model's output.
```

```python
actual_values = df['LoanDefault']
predicted_values = df['LoanDefault']

conf_matrix = confusion_matrix(actual_values, predicted_values)
conf_matrix
```

```python
#Heat map

sns.heatmap(
    conf_matrix,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['predicted 0', 'predicted 1'],
    yticklabels=['Actual 0', 'Actual 1']
) 

plt.xlabel('predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
